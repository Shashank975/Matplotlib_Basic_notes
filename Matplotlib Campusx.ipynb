```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Line Plot

```python
plt.style.use('fivethirtyeight')
```

```python
year=[2011,2012,2013,2014,2015,2016]
score=[90,88,80,25,92,95]
```

```python
plt.plot(year,score,color='red')
plt.title('Growth')
plt.xlabel('Years')
plt.ylabel('Score')
plt.show()
#line Plot
```

```python
batsman=pd.read_csv("sharma-kohli.csv")
batsman
```

```python
#fig=figure(figsize=(15,5))
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashdot',color='Blue',linewidth=2,marker="o",markersize=5,label='VK')
plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dotted',color="red",linewidth=2,marker="o",markersize=5,label='RS')
plt.title('Virat Kohli Vs Rohit Sharma')
plt.xlabel('Years')
plt.ylabel('Score')
plt.legend()
plt.show()
```

```python
year=[2011,2012,2013,2014,2015,2016,2017]
score=[90,88,80,25,92,95,50]
plt.plot(year,score,color='red',linestyle='dashdot',linewidth=2,marker="o",markersize=5)
plt.title('Growth')
plt.xlabel('Years')
plt.ylabel('Score')
plt.ylim()
plt.xlim()
plt.grid()
plt.show()#plt.show() -> is the main function we make the graph
```

# Scatter Plot

```python
x=np.linspace(-10,10,50)
y=10*x+3+np.random.randint(0,300,50)
plt.scatter(x,y)
```

```python
batter=pd.read_csv("batter.csv")
batter=batter.head(50)
```

```python
tips=sns.load_dataset('tips')
tips
```

```python

plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*5)
plt.title('Coustomer ')
plt.xlabel('Total bill In $')
plt.ylabel('Tips In $')
plt.show()
```

```python
plt.plot(tips['total_bill'],tips['tip'])
plt.title('Coustomer ')
plt.xlabel('Total bill In $')
plt.ylabel('Tips In $')
plt.show()
```

```python
#This is a faster Technique than Scatter itself.. Try to make scatter using plt.plot
plt.plot(tips['total_bill'],tips['tip'],'o')
plt.title('Coustomer ')
plt.xlabel('Total bill In $')
plt.ylabel('Tips In $')
plt.show()
```

# Bar Charts

```python
childerns=[60,30,20]
colors=['Red','Green','Blue']
```

```python
#Vertical Bar Chart
plt.bar(colors,childerns,color='black')
```

```python
#Horizontal Bar Charts
plt.barh(colors,childerns,color='black')
```

```python
season=pd.read_csv("batsman_season_record.csv")
season
```

```python
plt.bar(season['batsman'],season["2017"],color='yellow')
plt.bar(season['batsman'],season["2016"],bottom=season["2017"],color='green')
plt.bar(season['batsman'],season["2015"],bottom=(season["2016"]+season["2017"]),color='red')
plt.show()
```

```python
plt.bar(season['batsman'],season["2015"],color="black",alpha=0.3,label='2015')
plt.bar(season['batsman'],season["2016"],alpha=0.1,label='2016')
plt.bar(season['batsman'],season["2017"],alpha=0.2,label='2017')
plt.xticks(rotation="vertical")
plt.legend()
plt.show()
```
